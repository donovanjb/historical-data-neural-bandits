```python
import numpy as np

# Biased distribution: higher probability for smaller integers
probabilities = np.linspace(1, 0.1, 5)  # Example: probabilities for integers 0 to 4
probabilities /= probabilities.sum()  # Normalize to sum to 1
context = np.random.choice(5, size=2, p=probabilities)
context = context / 5  # Normalize to [0, 1]
context = np.insert(context, 0, 1.0)  # Add intercept term

np.round(probabilities, 4), context
```
